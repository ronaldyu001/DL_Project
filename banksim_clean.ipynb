```python
## Cleaning Summary

# The BankSim dataset was cleaned by removing extra quote marks from categorical values. 
# The dataset has no missing values, no infinite numeric values, and no duplicate rows. 
# The constant columns `zipcodeOri` and `zipMerchant` were removed because they contain only one unique value. 
# The final cleaned dataset contains 594,643 rows and 8 columns.
```

```python
import pandas as pd
import numpy as np
```

```python
banksim = pd.read_csv(
    "/Users/tejaljadhav/Desktop/UCD/Spring2026/Deep Learning/DL_Project/Dataset/bs140513_032310.csv"
)

banksim.head()
```

```python
banksim.shape
```

```python
# To remove extra quote marks.
banksim_clean = banksim.copy()

for col in banksim_clean.select_dtypes(include="object").columns:
    banksim_clean[col] = banksim_clean[col].str.strip().str.strip("'")

banksim_clean.head()
```

```python
# To check missing values.
banksim_clean.isnull().sum()
```

```python
# To check missing values.

banksim_clean.isnull().sum().sum()

```

```python
# To check infinite values.

numeric_cols = banksim_clean.select_dtypes(include=[np.number]).columns

np.isinf(banksim_clean[numeric_cols]).sum()
```

```python
# To check duplicated

banksim_clean.duplicated().sum()
```

```python
# If duplicate exists.
banksim_clean = banksim_clean.drop_duplicates().reset_index(drop=True)
```

```python
# To check fraud distribution.
banksim_clean["fraud"].value_counts()
```

```python
banksim_clean["fraud"].value_counts(normalize=True) * 100
```

```python
# To check categorical values.
banksim_clean["age"].value_counts().sort_index()

```

```python
banksim_clean["gender"].value_counts()
```

```python
banksim_clean["category"].value_counts()
```

```python
# Check constant columns.
banksim_clean.nunique()

```

```python
# Since zipcodeOri and zipMerchant only have one value, dropping them.
banksim_clean = banksim_clean.drop(["zipcodeOri", "zipMerchant"], axis=1)
```
