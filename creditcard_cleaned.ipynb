```python
## Cleaning Summary

# The credit card dataset had no missing or infinite values, and all features were numeric. 
# Exact duplicate rows were removed, reducing the dataset from 284,807 to 283,726 rows. 
# The cleaned dataset now has 473 fraud cases out of 283,726 transactions.
```
